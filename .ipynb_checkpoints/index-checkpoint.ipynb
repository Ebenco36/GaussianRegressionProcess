```python
import numpy as np
import matplotlib.pyplot as plt

class GaussianProcessRegressorARD:
    def __init__(self, alpha=1e-5):
        self.alpha = alpha  # Ridge regularization parameter
        self.X_train = None
        self.y_train = None
        self.K_inv = None  # Inverse of the kernel matrix
        self.alpha_inv_y = None  # Precomputed alpha * y for efficient prediction

    def rbf_kernel(self, X1, X2, length_scales):
        # Radial Basis Function (RBF) kernel with ARD
        dist_sq = np.sum((X1 / length_scales) ** 2, axis=1)[:, np.newaxis] + np.sum((X2 / length_scales) ** 2, axis=1) - 2 * np.dot(X1 / length_scales, X2.T / length_scales.T)
        return np.exp(-0.5 * dist_sq)

    def fit(self, X, y, length_scales):
        self.X_train = X
        self.y_train = y

        # Compute kernel matrix
        K = self.rbf_kernel(X, X, length_scales) + self.alpha * np.eye(len(X))

        # Invert kernel matrix
        self.K_inv = np.linalg.inv(K)

        # Precompute alpha * y for efficient prediction
        self.alpha_inv_y = np.dot(self.K_inv, y)

    def predict(self, X, length_scales):
        # Compute kernel matrices
        k_star = self.rbf_kernel(X, self.X_train, length_scales)
        k_star_star = self.rbf_kernel(X, X, length_scales)

        # Mean prediction
        mean = np.dot(k_star, self.alpha_inv_y)

        # Covariance prediction
        cov = k_star_star - np.dot(k_star, np.dot(self.K_inv, k_star.T))

        return mean, cov

# Example usage:

# Generate some synthetic data for illustration
np.random.seed(42)
X = np.random.rand(20, 1) * 10
y = np.sin(X[:, 0]) + np.random.randn(20) * 0.1

# Instantiate and fit the GPR model
gpr_model = GaussianProcessRegressorARD()
length_scales = np.ones(X.shape[1])  # Initialize length scales
gpr_model.fit(X, y, length_scales)

# Make predictions
X_new = np.linspace(0, 10, 100).reshape(-1, 1)
mean, cov = gpr_model.predict(X_new, length_scales)

# Plot the results
plt.scatter(X, y, color='red', label='Training Data')
plt.plot(X_new, mean, color='blue', label='Mean Prediction')
plt.fill_between(X_new.flatten(), mean - np.sqrt(np.diag(cov)), mean + np.sqrt(np.diag(cov)), color='blue', alpha=0.2, label='Uncertainty')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Gaussian Process Regression with ARD')
plt.legend()
plt.show()
```
